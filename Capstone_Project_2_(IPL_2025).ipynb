## Feature engineering

Create relevant features from the existing dataframes that can be used to train a predictive model. This might involve calculating team statistics, player statistics, and historical match data.


**Reasoning**:
Calculate team-level statistics like average runs scored, average wickets taken, average runs conceded, average wickets lost, and winning percentage from the matches and deliveries dataframes. Store these statistics in dictionaries or dataframes for later use in feature engineering.



```python
# Calculate average runs scored by each team in first and second innings
team_avg_runs_innings = matches_df.groupby('team1')['first_ings_score'].mean().to_frame(name='avg_first_ings_score')
team_avg_runs_innings['avg_second_ings_score'] = matches_df.groupby('team2')['second_ings_score'].mean()

# Calculate average wickets taken by each team in first and second innings
team_avg_wickets_innings = matches_df.groupby('team1')['first_ings_wkts'].mean().to_frame(name='avg_first_ings_wkts')
team_avg_wickets_innings['avg_second_ings_wkts'] = matches_df.groupby('team2')['second_ings_wkts'].mean()

# Calculate average runs conceded by each team in first and second innings from deliveries_df
team_runs_conceded_innings = merged_df.groupby(['bowling_team', 'innings'])['runs_of_bat'].sum().unstack(fill_value=0)
team_balls_bowled_innings = merged_df.groupby(['bowling_team', 'innings']).size().unstack(fill_value=0)
team_overs_bowled_innings = team_balls_bowled_innings / 6
team_avg_runs_conceded_over_innings = team_runs_conceded_innings / team_overs_bowled_innings
team_avg_runs_conceded_innings = team_avg_runs_conceded_over_innings * 20 # Assuming 20 overs in an innings

# Calculate average wickets lost by each team in first and second innings from deliveries_df
team_wickets_lost_innings = merged_df[merged_df['player_dismissed'].notnull()].groupby(['batting_team', 'innings'])['player_dismissed'].count().unstack(fill_value=0)
team_innings_played_innings = merged_df.groupby(['batting_team', 'innings'])['match_no'].nunique().unstack(fill_value=0)
team_avg_wickets_lost_innings = team_wickets_lost_innings / team_innings_played_innings


# Calculate winning percentage of each team
team_wins = matches_df.groupby('match_winner').size().to_frame(name='wins')
team_matches_played = matches_df['team1'].value_counts().add(matches_df['team2'].value_counts(), fill_value=0).to_frame(name='matches_played')
team_winning_percentage = team_wins.join(team_matches_played, how='right').fillna(0)
team_winning_percentage['winning_percentage'] = (team_winning_percentage['wins'] / team_winning_percentage['matches_played']) * 100

display(team_avg_runs_innings)
display(team_avg_wickets_innings)
display(team_avg_runs_conceded_innings)
display(team_avg_wickets_lost_innings)
display(team_winning_percentage)
```

```python
# Drop columns with a large number of missing values
matches_features_df = matches_features_df.drop(columns=['wb_runs', 'wb_wickets', 'avg_runs_conceded_innings3', 'avg_runs_conceded_innings4', 'avg_wickets_lost_innings3', 'avg_wickets_lost_innings4', 'avg_runs_conceded_innings3_team2', 'avg_runs_conceded_innings4_team2', 'avg_wickets_lost_innings3_team2', 'avg_wickets_lost_innings4_team2'])

# Impute remaining numerical missing values with the mean
numerical_cols = matches_features_df.select_dtypes(include=['float64', 'int64']).columns
for col in numerical_cols:
    if matches_features_df[col].isnull().any():
        matches_features_df[col].fillna(matches_features_df[col].mean(), inplace=True)

# Identify categorical columns
categorical_cols = matches_features_df.select_dtypes(include=['object']).columns
print("\nCategorical columns:")
print(categorical_cols)
```
